##### Sélection des structures avec des bilans publics et - de 10M€ de CA

```python
# Filtrer : visibilité publique et CA ≤ 10 millions
df_filtre = df[
    (df["confidentiality"] == "Public") &
    (df["Chiffre_d_affaires"] <= 10_000_000)
]

# Vérifier combien de lignes restent
print(f"Nombre d'entreprises filtrées : {len(df_filtre)}")
```
