```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import json
```

```python
# Charger le dataset des entreprises fermées
df = pd.read_parquet('dataset_closed_final.parquet')

# Vérification rapide
print(f"Dataset fermé chargé : {df.shape[0]} lignes, {df.shape[1]} colonnes")
df.head()
```

```python
df.shape
```

```python
# Comptage du nombre de sociétés par âge
age_counts = df["age_estime"].value_counts().sort_index().reset_index()
age_counts.columns = ["age", "nombre_societes"]

# Barplot interactif
fig = px.bar(
    age_counts,
    x="age",
    y="nombre_societes",
    title="Répartition du nombre de sociétés par âge",
    labels={"age": "Âge (années)", "nombre_societes": "Nombre de sociétés"},
    text="nombre_societes",
    height=500,
    width=900
)

# Affichage du nombre sur les barres
fig.update_traces(textposition="outside")

# Affichage
fig.show()
```

##### Echantillonage

```python
# Tirage aléatoire de 10 000 lignes
df_sample = df.sample(n=10000, random_state=42).reset_index(drop=True)
```

```python
# Comptage du nombre de sociétés par âge
age_counts = df_sample["age_estime"].value_counts().sort_index().reset_index()
age_counts.columns = ["age", "nombre_societes"]

# Barplot interactif
fig = px.bar(
    age_counts,
    x="age",
    y="nombre_societes",
    title="Répartition du nombre de sociétés par âge",
    labels={"age": "Âge (années)", "nombre_societes": "Nombre de sociétés"},
    text="nombre_societes",
    height=500,
    width=900
)

# Affichage du nombre sur les barres
fig.update_traces(textposition="outside")

# Affichage
fig.show()
```
