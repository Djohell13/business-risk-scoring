#### Fusion des csv pour EDA

```python
import pandas as pd
```

```python
open = pd.read_parquet('dataset_open_final.parquet')

print(f"Dataset fermé chargé : {open.shape[0]} lignes, {open.shape[1]} colonnes")
open.head()
```

```python
closed = pd.read_parquet('dataset_closed_final.parquet')

print(f"Dataset fermé chargé : {closed.shape[0]} lignes, {closed.shape[1]} colonnes")
closed.head()
```

```python
full = pd.concat([open, closed], ignore_index=True)

print(full.shape)
print(full["fermeture"].value_counts())
```

```python
full.head()
```

##### Vérification des Nan
