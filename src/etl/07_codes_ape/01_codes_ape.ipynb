```python
import pandas as pd
from IPython.display import display
```

```python
file_path = "int_courts_naf_rev_2.xls"

ape_df = pd.read_excel(file_path, sheet_name="NAF rév. 2", engine="xlrd")

ape_df.columns = [c.strip().replace("\n","") for c in ape_df.columns]

# Vérifier les noms
print(ape_df.columns.tolist())
```

```python
ape_df = ape_df[ape_df["Code"].notna()]
ape_df = ape_df[~ape_df["Code"].str.contains("SECTION")]

# On garde uniquement le code et le libellé complet
ape_df = ape_df[["Code", "Intitulés de la  NAF rév. 2, version finale"]]
ape_df.columns = ["code_ape", "libelle_ape"]

# On réinitialise l'index pour l'affichage
ape_df = ape_df.reset_index(drop=True)

# Vérifions le résultat
ape_df.head(100)
```
