```python
import json
```

```python
# Charger les rÃ©sultats existants
with open("siren_attachments.json", "r") as f:
    results = json.load(f)

total_siren = len(results)

bilans_total = 0
bilans_avec_bilan_saisi = 0
bilans_exploitables = set()

for siren, attachments in results.items():
    for bilan in attachments.get("bilans", []):
        bilans_total += 1

        if bilan.get("bilanSaisi"):
            bilans_avec_bilan_saisi += 1
            bilans_exploitables.add((siren, bilan.get("numChrono")))

print(f"âœ… SIREN rÃ©cupÃ©rÃ©s : {total_siren}")
print(f"ðŸ“„ Bilans rÃ©cupÃ©rÃ©s (bruts) : {bilans_total}")
print(f"ðŸ“„ Bilans avec bilanSaisi : {bilans_avec_bilan_saisi}")
print(f"ðŸ“„ Bilans exploitables (clÃ© siren+numChrono) : {len(bilans_exploitables)}")
```
